```python
#import module
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import wbdata
import datetime
import plotly.express as px
```

```python
#Read the dataset
df = pd.read_csv('NETFLIX MOVIES AND TV SHOWS CLUSTERING.csv')
# Filter for release year greater than 2019
filtered_df = df[df['release_year'] > 2019]
# Select the 'country' column from the filtered DataFrame
country_data = filtered_df[['country']]
country_data.head()
```

```python
#Bar plot show
country_data = filtered_df['country'] # Extract the 'country' column as a Series
# count the frequency of each country
country_counts = country_data.value_counts()
# plot the bar chart
plt.bar(country_counts.index, country_counts.values, color='orange')
plt.title('Bar Plot of Release Year for India (After 2019)')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.show()


```

```python
#Bar plot show
country_data = filtered_df['country'] # Extract the 'country' column as a Series
# count the frequency of each country
country_counts = country_data.value_counts()
# plot the histogram chart
plt.hist(country_counts.values,bins=100, edgecolor='orange')
plt.title('Histogram of Release Year (After 2019)')
plt.xlabel('Frequencies of countries')
plt.ylabel('Number of countries')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.show()
```

```python
import plotly.express as px
df = pd.read_csv('NETFLIX MOVIES AND TV SHOWS CLUSTERING.csv')
fig=px.scatter(df,
              x='release_year',
              y='rating',
              color='rating',
              animation_frame='country',
              title='Rating of Dramas show over the years')
fig.update_layout(height=400, width=500)
fig.show()
```
